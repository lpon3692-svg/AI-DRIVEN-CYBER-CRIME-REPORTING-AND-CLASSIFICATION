```python
import pandas as pd
import random
import re
import nltk
import numpy as np
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Download stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords

# =========================================================
# PART 1: ALL CYBERCRIME CATEGORIES (20+)
# =========================================================

categories_keywords = {

    "Phishing": ["OTP", "KYC", "bank verification", "UPI link", "lottery", "refund"],
    "Financial Fraud": ["credit card", "debit card", "UPI", "loan app", "ATM"],
    "Identity Theft": ["Aadhaar", "PAN", "SIM card", "Gmail", "Facebook"],
    "Cyberbullying": ["Instagram", "WhatsApp", "Facebook", "college group"],
    "Malware Attack": ["ransomware", "trojan", "virus", "spyware"],

    "Online Shopping Fraud": ["fake product", "ecommerce", "order scam"],
    "Social Media Hacking": ["Instagram hacked", "account takeover"],
    "UPI Fraud": ["QR scam", "collect request", "UPI PIN"],
    "Cyber Stalking": ["harassment", "tracking", "threat messages"],
    "Data Breach": ["data leak", "server hack", "unauthorized access"],

    "Email Spoofing": ["fake email", "spoofed email"],
    "Tech Support Scam": ["remote access", "fake support"],
    "Cryptocurrency Fraud": ["bitcoin scam", "crypto theft"],
    "Job Scam": ["fake job", "HR fraud"],
    "Loan App Fraud": ["loan harassment", "loan scam"],

    "Online Gaming Fraud": ["game hack", "free coins"],
    "SIM Swap Fraud": ["sim swap", "otp hijack"],
    "Fake News Scam": ["fake news", "rumor"],
    "Deepfake Scam": ["deepfake", "voice clone"],
    "Romance Scam": ["dating scam", "love fraud"]
}

# =========================================================
# PART 2: DATASET GENERATION (10000+)
# =========================================================

templates = [
    "I received a fake {item} message",
    "Fraud happened using {item}",
    "My account hacked via {item}",
    "Suspicious activity in {item}",
    "Unauthorized access in {item}",
    "Scam related to {item}",
    "Fake alert about {item}",
    "Money lost due to {item}"
]

data = []

for cat, items in categories_keywords.items():
    for _ in range(500):   # 20 categories × 500 = 10,000
        complaint = random.choice(templates).replace("{item}", random.choice(items))
        data.append([complaint, cat])

df = pd.DataFrame(data, columns=["complaint", "category"])
print("Dataset Size:", len(df))

# =========================================================
# PART 3: TEXT CLEANING
# =========================================================

stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    text = text.lower()
    words = [w for w in text.split() if w not in stop_words]
    return " ".join(words)

df['cleaned'] = df['complaint'].apply(clean_text)

# =========================================================
# PART 4: ML MODEL
# =========================================================

vectorizer = TfidfVectorizer(max_features=10000, ngram_range=(1,2))
X = vectorizer.fit_transform(df['cleaned'])
y = df['category']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression(max_iter=3000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("\nAccuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# =========================================================
# PART 5: RECOMMENDATIONS (SAMPLE SHORT VERSION)
# =========================================================

recommendations = {cat: [f"Follow safety step {i}" for i in range(1,21)] for cat in categories_keywords.keys()}

# =========================================================
# PART 6: SEVERITY + CONFIDENCE
# =========================================================

def severity(text):
    if "money" in text:
        return "High"
    elif "hack" in text:
        return "Medium"
    else:
        return "Low"

# =========================================================
# PART 7: CYBERCRIME TREND MODEL
# =========================================================

rate_data = {
    "Year":[2015,2016,2017,2018,2019,2020,2021,2022,2023,2024],
    "Users":[300,350,400,450,500,550,600,650,700,750],
    "Transactions":[10,20,30,50,80,120,160,200,250,300],
    "Cases":[300,450,600,800,1200,1800,2500,3300,4200,5200]
}

df_rate = pd.DataFrame(rate_data)

lr = LinearRegression()
lr.fit(df_rate[['Year','Users','Transactions']], df_rate['Cases'])

# =========================================================
# PART 8: REAL-TIME SYSTEM
# =========================================================

print("\n=== CYBERCRIME AI SYSTEM ===")

while True:
    text = input("\nEnter complaint (exit to quit): ")

    if text.lower() == "exit":
        break

    cleaned = clean_text(text)
    vec = vectorizer.transform([cleaned])

    pred = model.predict(vec)[0]
    prob = model.predict_proba(vec).max()

    print("\nPredicted Category:", pred)
    print("Confidence:", round(prob*100,2), "%")
    print("Severity:", severity(text))

    print("\nTop Recommendations:")
    for r in recommendations[pred][:10]:
        print("✔", r)

# =========================================================
# PART 9: FUTURE PREDICTION
# =========================================================

future = pd.DataFrame({
    "Year":[2025],
    "Users":[800],
    "Transactions":[350]
})

pred_2025 = int(lr.predict(future)[0])

print("\nPredicted 2025 Cases:", pred_2025)

plt.figure()
years = df_rate['Year'].tolist() + [2025]
cases = df_rate['Cases'].tolist() + [pred_2025]

plt.plot(years, cases, marker='o')
plt.title("Cybercrime Prediction")
plt.xlabel("Year")
plt.ylabel("Cases")
plt.grid()
plt.show()
```
