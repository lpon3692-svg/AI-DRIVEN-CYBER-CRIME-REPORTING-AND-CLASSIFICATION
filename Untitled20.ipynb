```python
import pandas as pd
import random
import re
import nltk
import numpy as np
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Download stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords

# =========================================================
# PART 1: LARGE DATASET (4800 RECORDS)
# =========================================================

categories_keywords = {
    "Phishing": ["OTP", "KYC", "bank verification", "UPI link", "lottery", "refund email"],
    "Financial Fraud": ["credit card", "debit card", "UPI", "loan app", "ATM fraud"],
    "Identity Theft": ["Aadhaar", "PAN", "SIM card", "email hacked", "fake profile"],
    "Cyberbullying": ["Instagram abuse", "WhatsApp harassment", "Facebook comments"],
    "Malware Attack": ["ransomware", "trojan", "virus", "spyware", "malicious app"],
    "Online Shopping Fraud": ["fake Amazon", "fake Flipkart", "online order scam"],
    "Job Scam": ["fake job offer", "work from home scam", "HR fraud"],
    "Social Media Hack": ["Instagram hacked", "Facebook hacked", "account takeover"],
    "UPI Fraud": ["Google Pay fraud", "PhonePe scam", "QR code scam"],
    "Cyber Stalking": ["threat messages", "online stalking", "unknown calls"],
    "Data Breach": ["data leak", "password leak", "database hacked"],
    "Fake News / Impersonation": ["fake profile", "impersonation", "deepfake"]
}

templates = [
    "I got a suspicious message about {item}.",
    "Someone misused my {item} details.",
    "Fraud transaction happened using {item}.",
    "My account was hacked through {item}.",
    "I clicked a fake link related to {item}.",
    "Unauthorized login detected in {item}.",
    "Scammer contacted me regarding {item}.",
    "Money was stolen via {item}.",
    "Received fake notification about {item}.",
    "My personal data exposed through {item}.",
    "I lost money because of {item}.",
    "Cybercriminal used {item} to attack me."
]

data = []
for cat, items in categories_keywords.items():
    for _ in range(400):  # 12 × 400 = 4800
        complaint = random.choice(templates).replace("{item}", random.choice(items))
        data.append([complaint, cat])

df = pd.DataFrame(data, columns=["complaint", "category"])
print("Dataset Size:", len(df))

# Save dataset
df.to_csv("cybercrime_dataset_4800.csv", index=False)

# =========================================================
# PART 2: TEXT CLEANING
# =========================================================

stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    text = text.lower()
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)

df['cleaned'] = df['complaint'].apply(clean_text)

# =========================================================
# PART 3: TRAIN MODEL
# =========================================================

vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X = vectorizer.fit_transform(df['cleaned'])
y = df['category']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression(max_iter=3000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nReport:\n", classification_report(y_test, y_pred))

# =========================================================
# PART 4: RECOMMENDATION ENGINE
# =========================================================

recommendations = {
    "Phishing": ["Avoid clicking unknown links","Verify email sender","Enable 2FA","Report phishing","Do not share OTP"],
    "Financial Fraud": ["Contact bank immediately","Block card","Enable alerts","Avoid sharing PIN","Monitor account"],
    "Identity Theft": ["Change passwords","Enable 2FA","Report identity theft","Monitor accounts","Secure documents"],
    "Cyberbullying": ["Block user","Report abuse","Save evidence","Talk to authority","Restrict privacy"],
    "Malware Attack": ["Install antivirus","Scan system","Avoid unknown downloads","Update software","Disconnect internet"],
    "Online Shopping Fraud": ["Use trusted sites","Check reviews","Avoid advance payment","Verify seller","Report fraud"],
    "Job Scam": ["Verify company","Avoid paying fees","Check official website","Report scam","Do not share documents"],
    "Social Media Hack": ["Reset password","Enable 2FA","Logout all devices","Report hack","Check activity logs"],
    "UPI Fraud": ["Do not scan unknown QR","Verify receiver","Enable PIN security","Report fraud","Use official apps"],
    "Cyber Stalking": ["Block contact","Report police","Avoid sharing info","Change privacy settings","Save evidence"],
    "Data Breach": ["Change passwords","Enable alerts","Monitor accounts","Use strong passwords","Enable encryption"],
    "Fake News / Impersonation": ["Verify source","Report fake account","Avoid sharing","Check authenticity","Stay informed"]
}

def generate_recommendation(crime):
    return random.sample(recommendations[crime], 5)

# =========================================================
# PART 5: CYBERCRIME PREDICTION MODEL
# =========================================================

rate_data = {
    "Year": [2015,2016,2017,2018,2019,2020,2021,2022,2023,2024],
    "Users": [320,350,380,410,450,500,550,600,650,700],
    "Transactions": [10,15,20,30,50,80,120,160,210,270],
    "Cases": [300,420,550,700,950,1300,1900,2600,3400,4300]
}

df_rate = pd.DataFrame(rate_data)

Xr = df_rate[['Year','Users','Transactions']]
yr = df_rate['Cases']

reg = LinearRegression()
reg.fit(Xr, yr)

def risk(c):
    if c < 2000: return "Low"
    elif c < 4000: return "Medium"
    else: return "High"

# =========================================================
# PART 6: USER INTERFACE
# =========================================================

print("\n=== AI CYBERCRIME SYSTEM ===")

while True:
    text = input("\nEnter complaint (exit to stop): ")

    if text.lower() == "exit":
        break

    cleaned = clean_text(text)
    vec = vectorizer.transform([cleaned])
    pred = model.predict(vec)[0]

    print("\nPredicted Crime:", pred)

    print("\nRecommended Actions:")
    for i, act in enumerate(generate_recommendation(pred), 1):
        print(f"{i}. {act}")

    try:
        year = int(input("\nEnter Year: "))
        users = int(input("Internet Users (Million): "))
        trans = int(input("Transactions (Billion): "))

        future = pd.DataFrame({
            "Year":[year],
            "Users":[users],
            "Transactions":[trans]
        })

        cases = int(reg.predict(future)[0])
        print("Predicted Cases:", cases)
        print("Risk Level:", risk(cases))

    except:
        print("Invalid input!")

# =========================================================
# PART 7: GRAPH
# =========================================================

future_2025 = pd.DataFrame({"Year":[2025],"Users":[750],"Transactions":[340]})
pred_2025 = int(reg.predict(future_2025)[0])

years = df_rate['Year'].tolist() + [2025]
cases = df_rate['Cases'].tolist() + [pred_2025]

plt.plot(years, cases, marker='o')
plt.title("Cybercrime Prediction")
plt.xlabel("Year")
plt.ylabel("Cases")
plt.grid()
plt.show()
```
